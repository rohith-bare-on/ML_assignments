```python
# https://drive.google.com/file/d/1eUzzlSsq9XgPYlf2dN7hfUzk8vBJnwhq/view

import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D 
from sklearn import decomposition 
from sklearn import datasets 


import pandas as pd
```

#### In this assignment students have to transform iris data into 3 dimensions  and plot a 3d chart with transformed dimensions and colour each data  point with specific class. 

```python
iris_data = datasets.load_iris()
iris_data
```

```python
iris = pd.DataFrame(data= iris_data['data'],columns = iris_data['feature_names']) #get data

iris['species'] = iris_data['target'] # add species columns which has target numbers
iris['species'] = iris_data['target_names'][iris['species']] # replace with species_names

iris
```

```python
fig = plt.figure(figsize=(8,8))
ax = Axes3D(fig,elev=10)

ax.set_title("Iris Dataset",fontsize=30)
ax.set_xlabel("PC1",fontsize=20)
ax.set_ylabel("PC2",fontsize=20)
ax.set_zlabel("PC3",fontsize=20)

iris_reduced = decomposition.PCA(n_components=3).fit_transform(iris.iloc[:,0:4])

ax.scatter(iris_reduced[:, 0], iris_reduced[:, 1], iris_reduced[:, 2], c=iris_data['target'])
```
